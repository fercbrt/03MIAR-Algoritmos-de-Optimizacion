# Encontrar los dos puntos ms cercanos

Desarrollar un algoritmo que, dado un conjunto de puntos, encuentre los dos ms cercanos de una y dos dimensiones.

Se generan dos conjuntos de datos en una y dos dimensiones respectivamente con la ayuda de la librera random

```python
import random

LISTA_1D = [random.randrange(1, 10000) for x in range(1000)]

LISTA_2D = [(random.randrange(1, 10000), random.randrange(1, 10000)) for x in range(1000)]
print(LISTA_2D)
```

## Conjunto uni-dimensional

El conjunto uni-dimensional objeto de prueba para los algoritmos implementados a continuacin se trata de un vector de mil nmeros pseudoaleatorios comprendidos entre 1 y 10000

### Fuerza Bruta

En primer lugar se lleva a cabo la implementacin de un algoritmo por fuerza bruta para aquellos dos puntos ms cercanos entre s.

```python
def fuerza_bruta_vector(lista):
  min_dist = float("inf")
  min_puntos = None

  n = len(lista)
  for i in range(n):
    for j in range(i + 1, n):
      dist = abs(lista[i] -  lista[j])
      if dist < min_dist:
        min_dist = dist
        min_puntos = (lista[i], lista[j])

  return min_puntos, min_dist
```

```python
print("Fuerza bruta para una dimensin", fuerza_bruta_vector(LISTA_1D))
```

El algoritmo utiliza dos bucles anidados que recorren la lista de puntos. El bucle externo se ejecuta $n$ veces y el bucle interno se ejecuta $(n-1), (n-2), ..., 1$ veces, lo que resulta en:

$$\sum_{i=1}^{n-1} i = \frac{n(n-1)}{2} \approx O(n^2)$$

### Divide y Vencers

Se implementa un algoritmo basado en la tcnica divide y vencers para abordar el problema.

```python
def divide_venceras_vector(lista):
    if len(lista) <= 3:
        return fuerza_bruta_vector(lista)

    lista_ordenada = sorted(lista)

    medio = len(lista_ordenada) // 2
    izquierda = lista_ordenada[:medio]
    derecha = lista_ordenada[medio:]

    min_izq, dist_izq = divide_venceras_vector(izquierda)
    min_der, dist_der = divide_venceras_vector(derecha)

    if dist_izq < dist_der:
        min_dist = dist_izq
        min_numeros = min_izq
    else:
        min_dist = dist_der
        min_numeros = min_der

    punto_medio = lista_ordenada[medio]

    for i in range(medio - 1, -1, -1):
        if punto_medio - lista_ordenada[i] >= min_dist:
            break
        for j in range(medio, len(lista_ordenada)):
            if lista_ordenada[j] - punto_medio >= min_dist:
                break
            dist = abs(lista_ordenada[i] - lista_ordenada[j])
            if dist < min_dist:
                min_dist = dist
                min_numeros = (lista_ordenada[i], lista_ordenada[j])

    return min_numeros, min_dist
```

```python
print("Divide y vencers para una dimensin", divide_venceras_vector(LISTA_1D))
```

El algoritmo se descompone en las siguientes operaciones:
- Ordenamiento inicial con coste $O(n \log n)$
- El problema se divide en dos subproblemas de tamao $n/2$
- Verificacin de puntos en la frontera $O(n)$

La relacin de recurrencia es: $T(n) = 2T(n/2) + O(n)$

Aplicando el Teorema Maestro, esto resulta en $O(n \log n)$

## Conjunto bi-dimensional

El conjunto bi-dimensional objeto de prueba para los algoritmos implementados a continuacin se trata de un vector de mil pares de nmeros pseudoaleatorios comprendidos entre 1 y 10000

Para calcular la distancia entre dos puntos dentro de una matriz, obtenemos la distancia eucldea mediante la implementacin de la siguiente funcin.

```python
import math

def distancia(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)
```

### Fuerza bruta

```python
def fuerza_bruta_matriz(matriz):
  min_dist = float("inf")
  min_puntos = None

  n = len(matriz)
  for i in range(n):
    for j in range(i + 1, n):
      d = distancia(matriz[i], matriz[j])
      if d < min_dist:
        min_dist = d
        min_puntos = (matriz[i], matriz[j])

  return min_puntos, min_dist
```

```python
print("Fuerza bruta para dos dimensiones", fuerza_bruta_matriz(LISTA_2D))
```

Al igual que en el caso unidimensional, el algoritmo utiliza dos bucles anidados para comparar todos los pares posibles de puntos. La diferencia es que ahora se calcula la distancia euclidiana entre puntos en un plano.

- Bucle externo de $n$ iteraciones
- Bucle interno de $(n-1) + (n-2) + ... + 1 = \frac{n(n-1)}{2}$ comparaciones
- Clculo de distancia euclidiana con coste $O(1)$ por cada par

$$Total: \frac{n(n-1)}{2} \times O(1) = O(n^2)$$

### Divide y vencers

```python
def divide_venceras_matriz(matriz):
    if len(matriz) <= 3:
        return fuerza_bruta_matriz(matriz)

    matriz_ordenada = sorted(matriz, key=lambda p: p[0])

    medio = len(matriz_ordenada) // 2
    punto_medio_x = matriz_ordenada[medio][0]

    izquierda = matriz_ordenada[:medio]
    derecha = matriz_ordenada[medio:]

    min_izq, dist_izq = divide_venceras_matriz(izquierda)
    min_der, dist_der = divide_venceras_matriz(derecha)

    if dist_izq < dist_der:
        min_dist = dist_izq
        min_puntos = min_izq
    else:
        min_dist = dist_der
        min_puntos = min_der

    franja = [p for p in matriz_ordenada if abs(p[0] - punto_medio_x) < min_dist]

    franja_ordenada = sorted(franja, key=lambda p: p[1])

    for i in range(len(franja_ordenada)):
        j = i + 1
        while j < len(franja_ordenada) and (franja_ordenada[j][1] - franja_ordenada[i][1]) < min_dist:
            d = distancia(franja_ordenada[i], franja_ordenada[j])
            if d < min_dist:
                min_dist = d
                min_puntos = (franja_ordenada[i], franja_ordenada[j])
            j += 1

    return min_puntos, min_dist
```

```python
print("Divide y vencers para dos dimensiones", divide_venceras_matriz(LISTA_2D))
```

Para el caso bi-dimensional, el algoritmo implementado se basa en el algoritmo de Shamos y Hoey para el par de puntos ms cercanos 2D [1], del cul se deduce la siguiente complejidad.

1. **Ordenamiento por coordenada x:** $O(n \log n)$
2. **Divisin recursiva:** Se divide el plano en dos mitades, procesando recursivamente cada una
3. **Combinacin - Franja central:**
   - Filtrado de puntos en la franja con coste $O(n)$
   - Ordenamiento por coordenada y con coste $O(n \log n)$
   - Verificacin de pares en la franja con coste $O(n)$ (por propiedad geomtrica, solo se comprueban 7 puntos siguientes)

La relacin de recurrencia es: $T(n) = 2T(n/2) + O(n \log n)$

Sin embargo, con optimizaciones (pre-ordenamiento), se puede reducir al siguiente orden

$$T(n) = 2T(n/2) + O(n)$$

lo que resulta en $O(n \log n)$
